```python
from sqlalchemy import create_engine
import geopandas as gpd
from dotenv import load_dotenv
import os
```

```python
# Load database connection details from .env file
load_dotenv()

DB_HOST = os.getenv('DB_HOST')
DB_PORT = os.getenv('DB_PORT')
DB_NAME = os.getenv('DB_NAME')
DB_USER = os.getenv('DB_USER')
DB_PASSWORD = os.getenv('DB_PASSWORD')

# Create database connection URL
db_url = f'postgresql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}'
```

```python
# Connect to the PostGIS database
engine = create_engine(db_url)
```

```python
# List of tables to export
tables = [
    'phmsa_hl_incidents',
    'phmsa_gas_incidents',
    'usgs_earthquakes'
]
```

```python
# Export each table as a shapefile
for table in tables:
    print(f"Exporting {table}...")
    
    # Read PostGIS table into a GeoDataFrame
    gdf = gpd.read_postgis(f'SELECT * FROM {table}', engine, geom_col='geom')
    
    # Export GeoDataFrame to shapefile
    gdf.to_file(f"{output_dir}/{table}.shp", driver='ESRI Shapefile')
    
    print(f"{table} exported to {output_dir}/{table}.shp")

print("Shapefile export complete.")
```

```python
from sqlalchemy import text

with engine.connect() as conn:
    for table in tables:
        # Query the table metadata to get the geometry column name
        meta_query = f"SELECT f_geometry_column FROM geometry_columns WHERE f_table_name = '{table}'"
        result = conn.execute(text(meta_query)).fetchone()
        
        if result:
            geom_col = result[0]
            print(f"Table: {table}, Geometry Column: {geom_col}")
        else:
            print(f"No geometry column found for table: {table}")
```

```python
# Export each table as a shapefile
with engine.connect() as conn:
    for table in tables:
        print(f"Exporting {table}...")
        
        # Query the table metadata to get the geometry column name
        meta_query = f"SELECT f_geometry_column FROM geometry_columns WHERE f_table_name = '{table}'"
        result = conn.execute(text(meta_query)).fetchone()
        
        if result:
            geom_col = result[0]
            
            # Read PostGIS table into a GeoDataFrame
            gdf = gpd.read_postgis(f'SELECT * FROM {table}', conn, geom_col=geom_col)
            
            # Export GeoDataFrame to shapefile
            gdf.to_file(f"{output_dir}/{table}.shp", driver='ESRI Shapefile')
            
            print(f"{table} exported to {output_dir}/{table}.shp")
        else:
            print(f"Skipping table {table} due to missing geometry column")

print("Shapefile export complete.")
```
